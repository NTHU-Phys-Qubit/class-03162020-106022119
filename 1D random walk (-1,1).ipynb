```python
import numpy as np
import matplotlib.pyplot as plt

step=1000
position=np.zeros(step)

for i in range (1,step):
    position += np.random.random(step)*2-1

plt.figure(figsize=(8, 5))
plt.plot(position,'.')
plt.title("1D Random Walk (-1,1)")
plt.xlabel('Orders')
plt.ylabel('Position')
```

```python
plt.figure(figsize=(8, 5))
plt.hist(position,bins=50)
plt.title("1D Random Walk (-1,1)")
plt.xlabel('Position')
plt.ylabel('Times')
```

```python
runs = 1000
final = np.zeros(runs)

for j in range (1,runs):
    position =0
    for i in range (1,step):
        position += np.random.random(step)*2-1
    final[j] = position[i]
        
plt.figure(figsize=(8, 5))
plt.hist(final,bins=50)
plt.title("1D Random Walk after 1000 runs (-1,1)")
plt.xlabel('Position')
plt.ylabel('Times')
```
