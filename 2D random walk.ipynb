```python
import numpy as np
import matplotlib.pyplot as plt

step=1000
positionx=np.zeros(step)
positiony=np.zeros(step)
```

```python
for i in range (0,step):
    positionx += np.random.random(step)-0.5
    positiony += np.random.random(step)-0.5

plt.figure(figsize=(8, 5))
plt.plot(positionx,positiony,'.')
plt.title("2D Random Walk")
plt.xlabel('Orders')
plt.ylabel('Position')
```

```python
positionxy = np.sqrt(np.square(positionx)+np.square(positiony))
plt.figure(figsize=(8, 5))
plt.hist(positionxy,bins=50)
plt.title("2D Random Walk")
plt.xlabel('Position')
plt.ylabel('Times')
```

```python
runs = 1000
final = np.zeros(runs)

for j in range (1,runs):
    positionx = 0
    positiony = 0
    for i in range (0,step):
        positionx += np.random.random(step)-0.5
        positiony += np.random.random(step)-0.5
    positionxy = np.sqrt(np.square(positionx)+np.square(positiony))
    final[j] = positionxy[i]
        
plt.figure(figsize=(8, 5))
plt.hist(final,bins=50)
plt.title("2D Random Walk after 1000 runs")
plt.xlabel('Position')
plt.ylabel('Times')
```
